```python
import pandas as pd
import numpy as np
```

```python
train = pd.read_csv(f'{data_path}titanic_train.csv')
test = pd.read_csv(f'{data_path}titanic_test.csv')

display(train.head())
train.shape
```

```python
test.shape
```

```python
target = train.pop('survived')
target
```

```python
train.head()
train.shape
```

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
train['cabin']
```

```python
train.columns
cols = ['passengerid', 'pclass', 'name', 'gender', 'age', 'sibsp', 'parch','ticket','cabin', 'fare', 'embarked']
# cols = ['passengerid', 'pclass', 'name', 'gender', 'age', 'sibsp', 'parch','ticket', 'fare', 'embarked']
cols
```

```python
train_ft = train[cols].copy()
test_ft = test[cols].copy()
train_ft.shape, test_ft.shape
```

```python
age_mean = train_ft['age'].mean()
# fare_median = train_ft['fare'].median()
fare_median = train_ft['fare'].mean()
embarked_mode = train_ft['embarked'].mode()[0]
# cabin_mode = train_ft['cabin'].mode()[0]

age_mean

train_ft['age'] = train_ft['age'].fillna(age_mean)
train_ft['fare'] = train_ft['fare'].fillna(fare_median)
train_ft['embarked'] = train_ft['embarked'].fillna(embarked_mode)
# train_ft['cabin'] = train_ft['cabin'].mode()[0]
train_ft.isnull().sum()
train_ft.head()
train_ft.shape
```

```python
age_mean = test_ft['age'].mean()
# fare_median = test_ft['fare'].median()
fare_median = test_ft['fare'].mean()
embarked_mode = test_ft['embarked'].mode()[0]
# cabin_mode = test_ft['cabin'].mode()[0]

test_ft['age'] = test_ft['age'].fillna(age_mean)
test_ft['fare'] = test_ft['fare'].fillna(fare_median)
test_ft['embarked'] = test_ft['embarked'].fillna(embarked_mode)
# test_ft['cabin'] = test_ft['cabin'].mode()[0]

test_ft.isnull().sum()
test_ft.shape
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
cols
```

```python
enc = OneHotEncoder(handle_unknown = 'ignore')
enc.fit(train_ft[cols])
```

```python
enc.transform(train_ft[cols]).toarray()
```

```python
enc.get_feature_names_out()
```

```python
tmp = pd.DataFrame(
    enc.transform(train_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)

train_ft = pd.concat([train_ft, tmp], axis = 1).drop(columns = cols)
train_ft
```

```python
tmp = pd.DataFrame(
    enc.transform(test_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)

test_ft = pd.concat([test_ft, tmp], axis = 1).drop(columns = cols)
test_ft
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
train_ft.head()
train_ft.shape
```

```python
train_ft.columns
```

```python
train_ft[train_ft.columns] = scaler.fit_transform(train_ft)
train_ft
```

```python
test_ft[test_ft.columns] = scaler.transform(test_ft)
test_ft.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_valid, y_train, y_valid = train_test_split(train_ft, target, test_size = 0.3, random_state = 42)
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
# max_depth= , n_estimators
model = LogisticRegression(random_state = 42)
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_valid)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid, pred)
```
