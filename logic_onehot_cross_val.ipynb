```python
import pandas as pd
import numpy as np
```

```python
data_path = '/content/drive/MyDrive/data/'
```

```python
train_tita = pd.read_csv(f'{data_path}titanic_train.csv')
test_tita = pd.read_csv(f'{data_path}titanic_test.csv')
```

```python
train_tita.info()
```

```python
train_tita.isnull().sum()
```

```python
test_tita.isnull().sum()
```

```python
train_tita_cp = train_tita.copy()
test_tita_cp = test_tita.copy()
```

```python
train_tita_cp.shape, test_tita_cp.shape
```

```python
train_tita_cp.head()
```

```python
target = train_tita_cp.pop('survived')
target
```

```python
train_tita_cp.isnull().sum()
```

```python
test_tita_cp.isnull().sum()
```

```python
# cols = train_tita_cp.select_dtypes('object').columns
cols = train_tita_cp.columns
cols
```

```python
train_tita_ft = train_tita_cp[cols]
test_tita_ft = test_tita_cp[cols]
```

```python
age_mean = train_tita_ft['age'].mean()
fare_median = train_tita_ft['fare'].median()
cabin_mode = train_tita_ft['cabin'].mode()[0]
embarked_mode = train_tita_ft['embarked'].mode()[0]
```

```python
age_mean, fare_median, cabin_mode, embarked_mode
```

```python
train_tita_ft['age'] = train_tita_ft['age'].fillna(age_mean)
train_tita_ft['fare'] = train_tita_ft['fare'].fillna(fare_median)
train_tita_ft['cabin'] = train_tita_ft['cabin'].fillna(cabin_mode)
train_tita_ft['embarked'] = train_tita_ft['embarked'].fillna(embarked_mode)
```

```python
test_tita_ft['age'] = test_tita_ft['age'].fillna(age_mean)
test_tita_ft['fare'] = test_tita_ft['fare'].fillna(fare_median)
test_tita_ft['cabin'] = test_tita_ft['cabin'].fillna(cabin_mode)
test_tita_ft['embarked'] = test_tita_ft['embarked'].fillna(embarked_mode)
```

```python
train_tita_ft.isnull().sum().sum(), test_tita_ft.isnull().sum().sum()
```

```python
train_tita_ft.head()
```

```python
test_tita_ft.head()
```

```python
cols = train_tita_ft.select_dtypes('object').columns
cols
```

```python
train_tita_ft[cols]
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown= 'ignore')
```

```python
enc.fit(train_tita_ft[cols])
```

```python
enc.transform(train_tita_ft[cols]).toarray()
```

```python
enc.get_feature_names_out()
```

```python
tmp = pd.DataFrame(
    enc.transform(train_tita_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)

train_tita_ft = pd.concat([train_tita_ft, tmp], axis = 1).drop(columns = cols)
train_tita_ft
```

```python
tmp = pd.DataFrame(
    enc.transform(test_tita_ft[cols]).toarray(),
    columns = enc.get_feature_names_out()
)

test_tita_ft = pd.concat([test_tita_ft, tmp], axis = 1).drop(columns = cols)
test_tita_ft
```

```python
train_tita_ft.shape, test_tita_ft.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
train_tita_ft.columns
```

```python
train_tita_ft[train_tita_ft.columns]= scaler.fit_transform(train_tita_ft)
train_tita_ft
```

```python
test_tita_ft[test_tita_ft.columns]= scaler.transform(test_tita_ft)
test_tita_ft
```

```python
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(train_tita_ft, target, random_state = 42, test_size = 0.2)
```

```python
x_train.shape, x_valid.shape, y_train.shape, y_valid.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state = 42)
```

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_valid)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid, pred)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
cv = KFold(n_splits = 5, shuffle = True, random_state = 42)
```

```python
scores = cross_val_score(model, train_tita_ft, target, scoring = 'roc_auc', cv = cv, n_jobs = -1)
scores
```
